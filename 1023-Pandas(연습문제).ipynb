```python
# Pandas의 나머지 기능들에 대해서 알아보아요!
# Pandas가 가지는 집계함수가 있어요!
# Numpy의 집계함수와 비슷합니다.!!
# sum(), mean(), max(), count(), ...

# DataFrame을 가지고 알아보아요!
import numpy as np
import pandas as pd

data = [[2, np.nan],
        [7, -3],
        [np.nan, np.nan],
        [1,2]]

df = pd.DataFrame(data,
                  columns=['one','two'],
                  index=['a', 'b', 'c', 'd'])
display(df)

# 집계함수를 이용해 보아요!
# 가장 대표적인 함수는 sum(), mean()

# print(df.sum())
# one    10.0
# two    -1.0
# dtype: float64
# 우리 예상과는 다르게 몽땅더하는게 아니라..
# axis=0 으로 간주되어서 실행됩니다.
# numpy(ndarray)는 집계함수 이용시 만약 axis를 명시하지 않으면
# axis=None으로 간주하는데
# pandas의 DataFrame은 만약 axis를 명시하지 않으면
# axis=0으로 간주해요!(결과는 Series로 리턴되요!)

# print(df.mean(axis=0, skipna=True)) # NaN은 아예 무시되요!

# 만약 skipna를 False로 설정하면 어떻게 되나요?
# print(df.mean(axis=0, skipna=False))
# NaN은 실수로 간주하고 연산이 가능, 단 값을 알수 없어요!라는 의미이기
# 때문에 NaN과 연산하면 무조건 결과는 NaN으로 나와요!

# 이렇게 우리가 만든 DataFrame이 있는데
# 이 DataFrame안의 데이터를 사용하기위해서(데이터분석, 머신러닝의 입력)
# 이 두 경우 모두
# 일반적으로 NaN이 있으면 연산이 안됨
# 이런 NaN을 우리가 결치값 (결치, missing value)라고 부르는데
# 당연히 이런 값을 처리해줘야 해요
# 처리하는 방법은 크게 2가지
#1. 가장 좋은 방법은 NaN을 삭제
# 그런데 문제가 있다
# 우리가 일반적으로 데이터 분석이나 머신러닝 학습 데이터로 사용하 데이터는 이 데이터는 많으면 많을수록 좋아요
# 많으면 많을수록 좋다고 하는데 일반적으로 많다는 기준은?
# 사실 문제(domain)마다 다름 하지만 약 10만건 기준으로 잡음
# 10만개를 기준으록 그거보다 적으면 데이터가 작아요
# 10만개를 기준으로 그거보다 많으면 데이터가 많이 있음
#    현실적으로는 데이터를 그렇게 많이 확보하기가 쉽지 않아요!
#    그렇다는 얘기는 가뜩이나 데이터가 적다는 말이예요!
#    없는 와중에 NaN이 포함되어있는 행을 모두 지우면 당연하게 데이터량이 감소
#    결론적으로 데이터의 전체 양이 많은 경우 삭제하는게 가장 좋아요!
#    하지만 데이터량이 적은경우에는 삭제하지 않는게 좋아요!
# 2. 그럼 다른 방법이 있나요? imputation을 사용해요!
#    NaN으로 되어 있는 놈을 값을 변화시켜서 사용해요!
#    그럼 어떻게 변환시킬까요?
# 2-1. 집계함수를 이용(기본통계기법)해서 값을 대체 -> 평균
# 2-2. 머신러닝 기법을 이용해서 값을 대체 -> 추론된값을 이용

# ***pandas로 이런 결측치를 처리****
# 결측치삭제는 아래처럼 작업해요
#new_df = df.dropna(how='any')
#display(new_df)

#결측치를 이번에는 다른 값으로 대체
one_avg = df['one'].mean() #3.3333
two_min = df['two'].min() #-3

df['one'] = df['one'].fillna(value = one_avg)
df['two'] = df['two'].fillna(value = two_min)
display(df)
```

```python
# 이렇게 DataFrame을 아무생각없이 연결할 수 잇어요
# 행방향, 열방향으로 연결할 수 있습니다
# 단, 연결할 때 결과가 어떻게 나오는지는 주의

#이번에는 DataFrame을 연결하는 또 다른방법에 대해서 알아보아요
# join에 대해서 알아보아요

data1 = { "학번" : [1,2,3,4],
          "이름" : ["이지안","박동훈","이순신","강감찬"],
          "학년" : [2,4,1,3]}

data2 = { "학번" : [1,2,4,5],
          "학과" : ["CS","MATH","MATH","CS"],
          "학점" : [3.4,2.9,4.5,1.2]}

df1 =pd.DataFrame(data1)
df2 =pd.DataFrame(data2)

display(df1, df2)

# 두 개의 DataFrame을 결합(merge)해보아요
# 먼저 inner join에 대해서 알아보아요
result = pd.merge(df1,
                  df2,
                  on='학번',
                  how='inner')

display(result)


```

```python
# 이번에는...
# filter()라는 함수를 이용해보아요!
# 정수기 필터... 거름망처리를 하는거예요!
# filter를 이용할때는 람다식을 많이 이용해요!
# age열의 평균값이 30보다 작은 그룹만을 따로 선택하세요!
grouped = df.groupby(['class'])

# 3개의 group중 나이의 평균값이 30보다 작은 그룹만을 선택!
group_filter = grouped.filter(lambda x: x.age.mean() < 30)
display(group_filter)
```
