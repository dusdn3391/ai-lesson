```python
# Multiple Linear Regression을 구현해 봥요
# scikit-learn 으로 구현해서 특정 데이터로 prediction까지

import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.preprocessing import MinMaxScaler
# Ras Data Loading
original = pd.read_csv('/content/drive/MyDrive/빅데이터 과정 폴더/data/ozone/ozone.csv')
display(original)

df1 = original[['Solar.R','Wind','Temp','Ozone']]
display(df1)

# 데이터 전처리
# 결측치 처리
df2 = df1.dropna(how='any')
display(df2)
#이상치 처리
import numpy as np
import matplotlib.pyplot as plt
#정규화 처리
scaler_x = MinMaxScaler()
scaler_t = MinMaxScaler()
#scikit-learn구현은 정규화처리를 하지 않아도 되요
#내부적으로 정규화 처리가 진행되요
# scaler에게 최대값과 최소값을 알려줘요!
scaler_x.fit(df2.drop('Ozone', axis=1, inplace = False))
scaler_t.fit(df2['Ozone'].values.reshape(-1,1))
#Traning Data Set을 정의
# x_data = df2.drop('Ozone', axis=1, inplace=False).values
# t_data = df2['Ozone'].values.reshape(-1,1)
x_data = scaler_x.transform(df2.drop('Ozone',axis=1, inplace=False))
t_data = scaler_t.transform(df2['Ozone'].values.reshape(-1,1))

# Model 만들고 학습을 진행

sklearn_model =linear_model.LinearRegression()

sklearn_model.fit(x_data, t_data)

sklearn_model.predict([[130,10,70]])
#array([[25.74518848]])
```

```python
#위의 문제를 Tensorflow Keras로 구현해서
# [[130,10,70]] 데이터에 대한 오존량의 예측값 구해보기

#model 구현
import matplotlib.pyplot as plt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import SGD

keras_model = Sequential()

keras_model.add(Flatten(input_shape=(3,)))

output_layer = Dense(units=1,
                      activation='linear')
keras_model.add(output_layer)

keras_model.compile(optimizer=SGD(learning_rate=1e-2),
                    loss='mse')

keras_model.fit(x_data,
                t_data,
                epochs=1000)

my_data = np.array([[130, 10, 70]])
my_data_norm = scaler_x.transform(my_data)

# 이렇게 예측할때도 정규화된 데이터를 넣어서 예측을 수행해야 해요!
result = keras_model.predict(my_data_norm)
real_result = scaler_t.inverse_transform(result)

print(real_result)
#[[25.008686]
```
